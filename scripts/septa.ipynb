```python
import csv
import json
```

```python
base = r"C:\Users\telpi\Documents\GitHub\DDRCardDraw\src\songs\iidx.json"
tier = r"C:\Users\telpi\Documents\GitHub\DDRCardDraw\scripts\septa.csv"
meld = r"C:\Users\telpi\Documents\GitHub\DDRCardDraw\src\songs\iidx-septa.json"

with open(base, 'r', encoding='utf-8') as fp:
    base_data = json.load(fp)
```

```python
base_data["meta"]["usesDrawGroups"] = True
for s in base_data["songs"]:
    for c in s["charts"]:
        c["drawGroup"] = c["lvl"]
base_data["i18n"]["en"]["name"] = "IIDX AC (Low-Mid-High 12s)"
base_data["i18n"]["ja"]["name"] = "IIDX AC (Low-Mid-High 12s)"
```

```python
songlist = {s["name"]: s for s in base_data["songs"]}
```

```python
slot_map = {
    "a": "another",
    "n": "normal",
    "h": "hyper",
    "b": "beginner",
    "l": "leggendaria"
}
with open(tier, 'r', encoding='utf-8') as fp:
    reader = csv.DictReader(fp)
    for row in reader:
        if row["title"] in songlist:
            song = songlist[row["title"]]
            slot = slot_map[row["slot"]]
            tier_offset = int(row["tier"])
            chart = [c for c in song["charts"] if (c["style"] == "single" and c["diffClass"] == slot)]
            if len(chart) == 1:
                newDrawGroup = 12 + tier_offset
                chart[0]["drawGroup"] = newDrawGroup
                if newDrawGroup > base_data["meta"]["lvlMax"]:
                    base_data["meta"]["lvlMax"] = newDrawGroup
                print(f"--- Updated draw group for single {slot} chart in {row['title']} to {newDrawGroup}")
            else:
                print(f"!!! Couldn't find a single {slot} chart in {row['title']}")
        else:
            print(f"!!! Couldn't find {row['title']}")
```

```python
with open(meld, 'w', encoding='utf-8') as fp:
    json.dump(base_data, fp)
```
